### Inspecting R variables

VSCode's Jupyter Variables panel does not support R. Use the inline snippet below (re-runnable) to see what's in your R global env:

```python
.env_summary <- function() {
  objs <- ls(envir = .GlobalEnv)
  if (!length(objs)) return('empty')
  data.frame(
    name  = objs,
    class = sapply(objs, function(n) class(get(n, envir = .GlobalEnv))[1]),
    size  = sapply(objs, function(n) format(object.size(get(n, envir = .GlobalEnv)), units = 'auto'))
  )
}
.env_summary()
```
